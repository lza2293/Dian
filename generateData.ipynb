```python
import numpy as np
from skimage import io,transform
import os
```

# Generate data

```python
def generatedata(label_dir, imgs_dir):
    with open(label_dir,'r') as f:
        contents = f.readlines() 
    X = []
    y_ = []
    for content in contents:
        vals = content.split()
        curImg_dir = vals[0][:-2] + '00/'
        img = io.imread(imgs_dir + curImg_dir + vals[0] + '.jpeg')
        img = transform.resize(img, (192, 192))
        img = np.array(img)
        img = img[8:184,8:184,:]
        print(img.shape)
        
        X.append(img)
        cur_label = np.zeros(374)
        for i in range(1, len(vals)):
            cur_label[int(vals[i]) - 1] = 1
        y_.append(cur_label)
    X = np.array(X)
    y_ = np.array(y_)
    return X,y_
```

### generate training data

```python
label_dir = './labels/training_label'
imgs_dir = './images/'
X,y_ = generatedata(label_dir,imgs_dir)

np.save('./data/train_x.npy',X)
np.save('./data/train_y.npy', y_)
print(X.shape)
```

### generate test data

```python
label_dir = './labels/test_label'
imgs_dir = './images/'
X,y_ = generatedata(label_dir,imgs_dir)

np.save('./data/test_x.npy',X)
np.save('./data/test_y.npy', y_)
```

```python
import matplotlib.pyplot as plt

filename = './Labels.npy'
Y = np.load(filename,allow_pickle = True)

Y = np.ravel(Y)
err_matrix = np.zeros((374, 374), dtype = np.int32)
count = 0

for y in Y:
    for i in range(len(y)):
        for j in range((i+1),len(y)):
                err_matrix[y[i] - 1][y[j] - 1] = err_matrix[y[i] - 1][y[j] - 1] + 1
                count = count + 1
print(count)
print(err_matrix)
```
